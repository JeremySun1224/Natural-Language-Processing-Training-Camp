# Word2vec综述

- 做NLP，尤其是中文，要理解句子，首先要将句子切分成词语。举个例子：判断一个词的词性是动词还是明细，用机器学习的思路，我们有一系列样本$(x, y)$，这里$x$是词语，$y$是它们的词性，我们要构建$f(x) -> y$的映射，但这里的数学模型$f(比如神经网络、SVM)$只接受数值型输入，而NLP里的词语是人类的抽象总结，是符号形式的（比如中文、英文、拉丁文等），所以我们需要把它们转换成数值形式，或者说——嵌入到一个数学空间里。这种嵌入方式，就叫做词嵌入(Word Embedding)，其中Word2vec就是词嵌入的一种。

- 至于后面如何处理词向量，比如求余弦，这是下次解决的问题了。

## Using wikiextractor to extrac the wikipedia corpus

- 使用[Python Wikipedia Extractor](https://github.com/attardi/wikiextractor)抽取维基百科的内容

- 从下载的语料库**xml格式**的压缩包中抽取维基百科的内容，将里面的**繁体字转化为简体字**（这是由于维基百科中很多是繁体中文网页）。另外，在用语料库训练词向量之前需要对中文句子进行分词，这里采用**Jieba中文分词**工具附句子进行分词，并保存在**reduce_zhiwiki.txt**文档中。具体实现方式如下：

```python
from gensim.corpora import WikiCorpus
from langconv import *
import jieba
```

```python
def my_function():
    space = " "
    i = 0
    l = []
    zhwiki_name = "./data/zhwiki-20190720-pages-articles-multistream.xml.bz2"
    f = open("./data/reduce_zhiwiki.txt", "w", encoding="utf-8")
    wiki = WikiCorpus(zhwiki_name, lemmatize=False, dictionary={})  # 从xml文件中读出训练语料
    
    for text in wiki.get_texts():
        for temp_sentence in text:
            temp_sentence = Converter("zh-hans").convert(temp_sentence)  # 繁体中文转化为简体中文
            seg_list = list(jieba.cut(temp_sentence))  # 分词
            for temp_term in seg_list:
                l.append(temp_term)
        f.write(space.join(l) + "\n")
        l = []
        i = i + 1
        
        if (i % 200 == 0):
            print("Saved " + str(i) + " articles")
    f.close()

if __name__ == "__main__":
    my_function
```

```python
my_function()
```

- 分词后的文件截图如下所示，文件有1.24G。在一般的NLP处理中，会需要使用停用词等处理。由于Word2vec的算法依赖于上下文，而上下文有可能就是停用词。因此这里对于Word2vec可以不用停用词，具体情况根据自己的情况来定。

## Using Gensim to get word vectors

- 参考Gensim的文档和Kaggle的参考文档，获得词向量。注意，你要使用Jieba分词把维基百科的内容切成一个一个单词，然后存进新的文件中。最后，你需要使用Gensim的LineSentence这个类进行文件的读取，再之后开始训练词向量Model；

- 参考[Word2vec](https://radimrehurek.com/gensim/models/word2vec.html)文档；

- 在训练之前，我们需要了解Gensim里面的参数。在gensim中，Word2vec相关的API都在类gensim.models.word2vec.Word2Vec中。算法中需要注意的参数如下：<br><br/>
    - **sentences**：我们要分词的语料，可以使一个列表，或者从文件中遍历读出。<br><br/>
    - **size**：词向量的维度，默认值是100.这个维度的取值一般与我们的语料大小相关。如果是不大的语料，比如小于100M的文本语料，则使用默认值一般就可以了。如果是超大的语料，建议增大维度。<br><br/>
    - **window**：即词向量上下文的最大距离，这个参数在我们的算法原理篇中标记为：window越大，则和某一词较远的词也会产生上下文关系。默认值为5。在实际使用过程照片那个，可以根据实际的需求来动态调整这个window的大小。如果是小语料，这个值可以设置的更小。对于一般的语料，这个值推荐在[5, 10]之间。<br><br/>
    - **sg**：即我们Word2vec两个模型的选择了。如果是0，则是CBOW模型，是1则是Skip-Gram模型。默认是0，即CBOW模型。<br><br/>
    - **hs**：即我们Word2vec两个解法的选择，如果是0，则是Negative Sampling，是1的话并且负采样个数negative大于0，则是Hierarchical Softmax。默认是0，即Negative Sampling。<br><br/>
    - **negative**：即使用Negative Sampling时负采样的个数，默认是5.推荐在[3, 10]之间。这个参数在我们的算法原理篇中标记为neg。<br><br/>
    - **cbow_count**：仅用于CBOW在做投影的时候。为0则算法中的为上下文的词向量之和，为1则为上下文的词向量的平均值。在我们的原理篇中，是按照词向量的平均值来描述的。个人比较喜欢用平均值来表示，默认也是1，不推荐修改默认值。<br><br/>
    - **min_count**：需要计算词向量的最小词频。这个值可以去掉一些生僻的低频词。默认值是5，如果是小语料，可以调低这个值。<br><br/>
    - **alpha**：在随机梯度下降法中迭代的初始步长，默认是0.025。<br><br/>
    - **min_alpha**：由于算法在迭代的过程中逐渐减小步长，min_alpha给出了最小的迭代步长值。随机梯度下降中每轮的迭代步长可以由iter，alpha，min_alpha一起得出。这部分由于不是Word2vec算法的核心内容，因此在原理篇我们没有提到。对于大语料，需要对alpha，min_alpha，iter一起调参来选择合适的三个值。<br><br/>
    - **worker**：训练词向量使用时的线程数，默认为3。<br><br/>

- 这里计算词语“大数据”与“人工智能”之间的相似度会报错：not in vocabulary。这是由于这两个词不在训练的词向量里面，可能的原因如下：
    - 我们在训练词向量的时候，min_count最小词频默认为5，也就是分词后的语料中词频低于5的词语不会被训练。这里我么可以将词频设置小一些没比如min_count=2。语料越小，设置的词频应该越低，但训练的时间成本也会随之上升；
    - 可能语料里没有这个词，这个原因的可能性很小；
    - 可能繁体字转化为简体字的时候没有转化过来。

## Using visualization tools

### Word Cloud

```python
import logging
import numpy as np
import matplotlib.pyplot as plt
from gensim import models
from wordcloud import WordCloud
```

```python
font_path = r"C:\Windows\Fonts\simsun.ttc"
```

```python
# 获取一个圆形的mask
def get_mask():
    x, y = np.ogrid[:300, :300]
    mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
    mask = 255 * mask.astype(int)
    return mask
```

```python
# 绘制词云
def draw_word_cloud(word_cloud):
    wc = WordCloud(font_path=font_path, background_color="white", mask=get_mask())
    wc.generate_from_frequencies(word_cloud)
    # 隐藏x轴和y轴
    plt.axis("off")
    plt.imshow(wc, interpolation="bilinear")
    plt.show()
```

```python
# 测试
def test():
    logging.basicConfig(format="%(asctime)s: %(levelname)s: %(message)s", level=logging.INFO)  # logging.basicConfig打印日志
    model = models.Word2Vec.load("./model/zhiwiki_news.word2vec")
    # 输入一个词找出相似的前100个，默认是10
    one_corpus = ["弦续"]
    result = model.most_similar(one_corpus[0], topn=100)
    # 将返回的结果转换为字典，便于绘制词云
    word_cloud = dict()

    for sim in result:
        word_cloud[sim[0]] = sim[1]

    # 绘制词云
    draw_word_cloud(word_cloud)
```

```python
test()
```

## 使用t-SEN进行词向量的可视化

- [visualizing-word-vectors](https://www.kaggle.com/jeffd23/visualizing-word-vectors-with-t-sne)

```python
import pandas as pd
import matplotlib.pyplot as plt
import logging
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
from sklearn.manifold import TSNE
from matplotlib.font_manager import FontProperties
```

```python
pd.options.mode.chained_assignment = None
```

```python
font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=14)
```

```python
logging.basicConfig(format="%(asctime)s: %(levelname)s: %(message)s", level=logging.INFO)
```

```python
wiki_news = open("./data/reduce_zhiwiki.txt", "r", encoding="utf-8")
model = Word2Vec(LineSentence(wiki_news), sg=0, size=200, window=10, min_count=500, workers=6)
print("Model training has been completed.")

def tsne_plot(model):
    labels = []
    tokens = []
    
    for word in model.wv.vocab:
        tokens.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=40, n_components=2, init="pca", n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)
    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(16, 16))
    for i in range(len(x)):
        plt.scatter(x[i], y[i])
        plt.annotate(labels[i],
                    fontproperties=font,
                    xy=(x[i], y[i]),
                    xytext=(5, 2),
                    textcoords="offset points",
                    ha="right",
                    va="bottom")
    
    plt.show()

tsne_plot(model)
```

- 由于上述训练量太大，故重新找一篇其他文档进行词向量训练并可视化。

```python
# 数据预处理
import collections
import re
import jieba
```

```python
#读取数据
def read_data():
    """
    对要训练的文本进行处理，最后把文本的内容的所有词放在一个列表中
    """
    #读取停用词
    stop_words = []
    with open('./data/stopwords/中文停用词表.txt', "r", encoding="utf-8-sig") as f:
        line = f.readline()
        while line:
            stop_words.append(line[:-1])
            line = f.readline()
    stop_words = set(stop_words)
    print('停用词读取完毕，共{n}个词'.format(n=len(stop_words)))
 
    # 读取文本，预处理，分词，得到词典
    raw_word_list = []
    rules =  u"([\u4e00-\u9fff]+)"  #只提取中文
    pattern =  re.compile(rules)
    with open('./data/人生.txt',"r", encoding='gb18030') as f:
        lines = f.readlines()
        for line in lines:
            line = line.replace("\r","").replace("\n","").strip()
            if line == "" or line is None:
                continue
            line = ' '.join(jieba.cut(line))
            seg_list = pattern.findall(line)
            words_list = []
            for word in seg_list:
                if word not in stop_words:
                    words_list.append(word)
            with open('./data/分词后的人生.txt', 'a', encoding='gb18030') as ff:
                if len(words_list) > 0:
                    line = ' '.join(words_list) + "\n"
                    ff.write(line) # 词汇用空格分开
                    ff.flush()
                    raw_word_list.extend(words_list)
    return raw_word_list
```

```python
words = read_data()
print(len(words))
print(len(set(words)))
```
